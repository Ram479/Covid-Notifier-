```python
import requests
import time
from bs4 import BeautifulSoup
page = requests.get("https://www.worldometers.info/coronavirus/country/india/")
page1 = requests.get("https://www.worldometers.info/coronavirus/")
soup = BeautifulSoup(page.content, 'html.parser')
soup1 = BeautifulSoup(page1.content, 'html.parser')
value1=soup1.find_all(class_="maincounter-number")
value=soup.find_all( class_="maincounter-number")
k=value1[0].get_text()
print("Total Active Case of Corona in the world is : "+k)
p=value[0].get_text()
p=p.strip("\n")
p1=value[1].get_text()
p1=p1.strip("\n")
p2=value[2].get_text()
p2=p2.strip("\n")
title="COVID-19 Alert"
message=f"Till now, there are {p}active cases in india in which {p1} has been reported dead and {p2} has been recovered so far... So please stay home and stay safe.. "
from plyer import notification

if __name__=="__main__":
    while True:
        
        notification.notify(
        title=title,
        message=message,
        timeout=10)
    time.sleep(2)
```
